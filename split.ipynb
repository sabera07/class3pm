```python
from bs4 import BeautifulSoup
import requests
url= BeautifulSoup('https://www.sharesansar.com/live-trading','html.parser')
soup = requests.get(url)
print(soup)
```

```python
code = BeautifulSoup(soup.text,'lxml')
code = code.table
code
```

```python
for tag in code:
    print(tag.text)
    
```

```python
data = []
for tag in code:
    data.append(tag.text)
data
```

```python
data[1].split('\n')
```

```python
for i in data[3:]:
    print(i.split("\n\n\n                                        "))
```

```python
tags = code.find_all('tr')
tags
```

```python
data = []
for tag in tags:
    data.append(tag.find_all('td'))
data
```

```python
final_data = []
for i in data:
    k = []
    for j in i:
        z = j.text.replace('                                        ','').replace('\n','')
        k.append(z)
    final_data.append(k)
final_data = [['sn','symbol','ltp','point change','open','high','low','volume','close']] + final_data
```

```python
import csv
file = open('nepse.csv','w')
x = csv.writer(file)
for i in final_data:
    x.writerow(i)
file.close()
```
